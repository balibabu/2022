```python
import math
from matplotlib import pyplot as plt

x_values=[i for i in range(-10,11)]
xsinx=[(x*math.sin(x)) for x in x_values]
x2sinx=[(pow(x,2)*math.sin(x)) for x in x_values]
x3sinx=[(pow(x,3)*math.sin(x)) for x in x_values]
x4sinx=[(pow(x,4)*math.sin(x)) for x in x_values]

plt.plot(x_values,xsinx,label='xsinx',color='red',linestyle='solid')
plt.plot(x_values,x2sinx,label='x2sinx',color='GREEN',linestyle='solid')
plt.plot(x_values,x3sinx,label='x3sinx',color='blue',linestyle='solid')
plt.plot(x_values,x4sinx,label='x4sinx',color='black',linestyle='--')

plt.legend(loc=9)
plt.show()
```

```python
def sumOfDigit(num):
    sum_digits=0
    while(num):
        sum_digits+=num%10
        num//=10
    return sum_digits

lowerR=int(input('enter lower range '))
upperR=int(input('enter upper range '))

perfectSq=[]
numList=[]
for i in range(lowerR,upperR+1):
    a=pow(i,2)
    if a <= upperR and sumOfDigit(a)<10:
        perfectSq.append(a)
        numList.append(i)


plt.plot(perfectSq,numList,color='red')
plt.title('Perfect Squares')
plt.xlabel('squares --------->')
plt.ylabel('numbers ---------->')
plt.show()
```
